# scikit-learnの導入
機械学習ライブラリscikit-learnの使い方に慣れていきましょう。  
今回は、導入としてIrisの品種分類を行います。

## ● Iris dataset
Setosa、Versicolor、Versinicaの3つの品種について、以下をグラフ表示します。  
* Sepal length
* Sepal width
* Petal length
* Petal width

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Irisデータの読み込み
iris = datasets.load_iris()

# 各花のサイズ
iris_data = iris.data

# 散布図で表示
st_data = iris_data[:50]  # Setosa
vc_data = iris_data[50:100]  # Versicolor
vn_data = iris_data[100:150]  # Versinica

print(st_data, type(st_data))

plt.scatter(st_data[:, 0], st_data[:, 1], label="Setosa")  # Sepal lengthとSepal width
plt.scatter(vc_data[:, 0], vc_data[:, 1], label="Versicolor")  # Sepal lengthとSepal width
plt.scatter(vn_data[:, 0], vn_data[:, 1], label="Versinica")  # Sepal lengthとSepal width
plt.legend()
plt.xlabel("Sepal length (cm)")
plt.ylabel("Sepal width (cm)")
plt.show()

plt.scatter(st_data[:, 2], st_data[:, 3], label="Setosa")  # Petal lengthとPetal width
plt.scatter(vc_data[:, 2], vc_data[:, 3], label="Versicolor")  # Petal lengthとPetal width
plt.scatter(vn_data[:, 2], vn_data[:, 3], label="Versinica")  # Petal lengthとPetal width
plt.legend()
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
plt.show()
```

## ● サポートベクターマシン
scikit-learnには様々な機械学習のアルゴリズムが含まれています。  
今回は、サポートベクターマシンという手法によりIrisの品種分類を行います。  
サポートベクターマシンのインスタンスを生成し、`fit`メソッドを実行するのみで分類を行うことが可能です。

```python
from sklearn import datasets
from sklearn import svm

iris = datasets.load_iris()
# print(iris.target)  # 品種を表すラベル 0:Setosa、1:Versicolor、2:Versinica

clf = svm.SVC()  # サポートベクターマシーン
clf.fit(iris.data, iris.target)  # 訓練

# 品種の判定 (Sepal length, Sepal width, Petal length, Petal width)
print(clf.predict([[5.1, 3.5, 1.4, 0.1], [6.5, 2.5, 4.4, 1.4], [5.9, 3.0, 6.2, 2.0]]))
```

```python
# コード練習用
import pandas as pd
pred_list = []
for a in range(10):
    for b in range(10):
        for c in range(10):
            for d in range(10):
                para = [4+0.5*a, 2+0.5*b, 1+1.25*c, 0.25*d]
                pred = clf.predict([para])
                pred_list.append([4+0.5*a, 2+0.5*b, 1+1.25*c, 0.25*d, pred[0]])
pred_list = np.array(pred_list)
print(pred_list)

arr = np.random.rand(1000, 5)
arr[:, 0] = 4 + 4 * arr[:, 0]
arr[:, 1] = 2 + 2.5 * arr[:, 1]
arr[:, 2] = 1 + 6 * arr[:, 2]
arr[:, 3] = 2.5 * arr[:, 3]
arr[:, 4] = clf.predict(arr[:, 0:4])
print(arr)

# print(len(pred_list))
# print(pred_list[9999][0], pred_list[9999][1], pred_list[9999][2], pred_list[9999][3])

# 散布図でmatplotlib.pyplot表示

st_data = []
vc_data = []
vn_data = []

for sl, sw, pl, pw, pre in arr:
    if pre == 0:
        st_data.append([sl, sw, pl, pw, pre])
    elif pre == 1:
        vc_data.append([sl, sw, pl, pw, pre])
    else:
        vn_data.append([sl, sw, pl, pw, pre])
        
st_data = np.array(st_data)
vc_data = np.array(vc_data)
vn_data = np.array(vn_data)
        
print(st_data)
print(len(st_data))
print(len(vc_data))
print(len(vn_data))

plt.scatter(st_data[:, 0], st_data[:, 1], label="Setosa")  # Sepal lengthとSepal width
plt.scatter(vc_data[:, 0], vc_data[:, 1], label="Versicolor")  # Sepal lengthとSepal width
plt.scatter(vn_data[:, 0], vn_data[:, 1], label="Versinica")  # Sepal lengthとSepal width
plt.legend()
plt.xlabel("Sepal length (cm)")
plt.ylabel("Sepal width (cm)")
plt.show()

plt.scatter(st_data[:, 2], st_data[:, 3], label="Setosa")  # Petal lengthとPetal width
plt.scatter(vc_data[:, 2], vc_data[:, 3], label="Versicolor")  # Petal lengthとPetal width
plt.scatter(vn_data[:, 2], vn_data[:, 3], label="Versinica")  # Petal lengthとPetal width
plt.legend()
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
plt.show()
```
